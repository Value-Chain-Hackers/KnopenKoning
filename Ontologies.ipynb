```python
query = """
PREFIX iof_av: <http://example.org/iof-av#>
PREFIX skos: <http://www.w3.org/2004/02/skos/core#>
PREFIX dcterms: <http://purl.org/dc/terms/>

SELECT ?subject ?naturalLanguageDefinition ?synonym ?abbreviation ?example ?definition
WHERE {
  ?subject a ?type .
  OPTIONAL { ?subject iof_av:naturalLanguageDefinition ?naturalLanguageDefinition . }
  OPTIONAL { ?subject iof-av:abbreviation ?abbreviation . }
 ?subject iof-av:synonym ?synonym . 
  OPTIONAL { ?subject skos:example ?example . }
  OPTIONAL { ?subject skos:definition ?definition . }
}
"""
# Execute the query
results = graph.query(query)

# Print the results
for row in results:
    print(f"Subject: {row.subject}, synonym: {row.synonym}, Natural Language Definition: {row.naturalLanguageDefinition}, Example: {row.example}, Definition: {row.definition}")
```

```python

from langchain_community.llms.ollama import Ollama
from langchain_experimental.graph_transformers import LLMGraphTransformer
from langchain_core.documents import Document
from langchain_community.graphs import Neo4jGraph
import os
os.environ["NEO4J_URI"] = "bolt://localhost:7687"
os.environ["NEO4J_USERNAME"] = "sebdg"
os.environ["NEO4J_PASSWORD"] = "password"
docs = [
    Document(
        page_content="Choosing the right suppliers is crucial for a resilient supply chain, and AI can help by assessing risk and performance factors. AI tools analyze supplier data, including reliability, cost, quality, and compliance, to recommend the best partners for your business. This informed decision-making process ensures that you collaborate with suppliers who can meet your standards and contribute to a smooth-running supply chain.",
        metadata={"source": "book"},
    )]

llm=Ollama(model="mistral:latest")
llm_transformer = LLMGraphTransformer(llm=llm)

graph_documents = llm_transformer.convert_to_graph_documents(docs)


# Store to neo4j
graph = Neo4jGraph(database="test")
graph.add_graph_documents(
  graph_documents, 
  baseEntityLabel=True, 
  
  include_source=True
)
```
