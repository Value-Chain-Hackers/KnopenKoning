```python
def get_knowledge_from_wikipedia(search_term):
    import wikipedia
    wikipedia.set_lang("en")
    search_results = wikipedia.search(search_term)
    if not search_results:
        return None
    pages = []
    for result in search_results:
        try:
            page = wikipedia.page(result)
            pages.append(page.content)
        except Exception as e:
            continue
    return pages
```

```python
from langchain_community.llms.ollama import Ollama
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import StrOutputParser
from langchain_core.documents import Document
```

```python

summarize = PromptTemplate.from_template("""\
Given the following text, summarize it in a few sentences and takeaways.
Make a list of the key points and summarize them in a few sentences.
create a list of thoughts and ideas from the text as pathways to evaluate a company behavior accoring to the supply chain.

------
Context:
                                    
{chunk}

------
""")

llm = Ollama(model="phi3:latest", num_ctx=4096, num_predict=2048, temperature=0.1)
chain = summarize | llm | StrOutputParser()
chain.invoke({"chunk": get_knowledge_from_wikipedia("supply chain mapping")[:2000]})
```
